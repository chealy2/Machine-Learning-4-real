```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('Downloads/D3.csv')
data.head(4)
x1=data.values[:, [0]]
x2=data.values[:, [1]]
x3=data.values[:, [2]]
y=data.values[:, [3]]
m=len(y)
print(m)
```

```python
X_0 = np.ones((m, 1))
X_1 = x1.reshape(m, 1)
X_2 = x2.reshape(m, 1)
X_3 = x3.reshape(m, 1)

Xs1 = np.hstack((X_0, X_1))
Xs2 = np.hstack((X_0, X_2))
Xs3 = np.hstack((X_0, X_3))

theta = np.random.randn(2,1)
```

```python
def compute_cost_input1(Xs1, y, theta):
    predictions1 = Xs1.dot(theta) 
    errors1 = np.subtract(predictions1, y)   
    sqrErrors1 = np.square(errors1)   
    J1 = 1 / (2 * m) * np.sum(sqrErrors1) 

    return J1
```

```python
def compute_cost_input2(Xs2, y, theta):
    predictions2 = Xs2.dot(theta) 
    errors2 = np.subtract(predictions2, y)   
    sqrErrors2 = np.square(errors2)   
    J2 = 1 / (2 * m) * np.sum(sqrErrors2) 
    
    return J2
```

```python
def compute_cost_input3(Xs3, y, theta):
    predictions3 = Xs3.dot(theta) 
    errors3 = np.subtract(predictions3, y)   
    sqrErrors3 = np.square(errors3)   
    J3 = 1 / (2 * m) * np.sum(sqrErrors3) 
    
    return J3
```

```python
cost1 = compute_cost_input1(Xs1, y, theta) 
print('The cost for given values of theta_0 and theta_1 =', cost1) 
theta = np.random.randn(2,1)

cost2 = compute_cost_input2(Xs2, y, theta) 
print('The cost for given values of theta_0 and theta_1 =', cost2) 
theta = np.random.randn(2,1)

cost3 = compute_cost_input3(Xs3, y, theta) 
print('The cost for given values of theta_0 and theta_1 =', cost3) 
```

```python
def gradient_descent(Xs1,y,theta,alpha,iterations):

    cost_history1 = np.zeros(iterations)

    for it in range(iterations):
        
        prediction = np.dot(Xs1,theta)
        
        theta = theta -(1/m)*alpha*( Xs1.T.dot((prediction - y)))
        
        cost_history1[it]  = compute_cost_input1(Xs1, y, theta) 
        
    return theta, cost_history1
```

```python
def gradient_descent2(Xs2,y,theta,alpha,iterations):

    cost_history2 = np.zeros(iterations)

    for it in range(iterations):
        
        prediction = np.dot(Xs2,theta)
        
        theta2 = theta2 -(1/m)*alpha*( Xs2.T.dot((prediction - y)))
        
        cost_history1[it]  = compute_cost_input1(Xs2, y, theta) 
        
    return theta2, cost_history2
```

```python
def gradient_descent3(Xs3,y,theta,alpha,iterations):

    cost_history3 = np.zeros(iterations)

    for it in range(iterations):
        
        prediction = np.dot(Xs3,theta)
        
        theta3 = theta3 -(1/m)*alpha*( Xs3.T.dot((prediction - y)))
        
        cost_history1[it]  = compute_cost_input1(Xs3, y, theta) 
        
    return theta3, cost_history3
```

```python
theta = np.random.randn(2,1)
iterations=1500;
alpha=0.01;

theta, cost_history1 = gradient_descent(Xs1,y,theta,alpha,iterations)
print('Final value of theta =',theta)
print('cost_history =',cost_history1)
```

```python
plt.scatter(Xs1[:,1], y, color='red', marker= '+', label= 'Training Data') 
plt.plot(Xs1[:,1],Xs1.dot(theta), color='green', label='Linear Regression')  
plt.rcParams["figure.figsize"] = (10,6) 
plt.grid() 
plt.xlabel('Xs1') 
plt.ylabel('y') 
plt.title('Linear Regression Fit') 
plt.legend()
```

```python
plt.plot(range(1, iterations + 1),cost_history1, color='blue') 
plt.rcParams["figure.figsize"] = (10,6) 
plt.grid() 
plt.xlabel('Number of iterations') 
plt.ylabel('Cost (J1)') 
plt.title('Convergence of gradient descent')
```
